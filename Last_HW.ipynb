```python
from urllib.request import urlopen
from random import choice

# Function to create a dictionary where each word is a key, and the corresponding value
# is a list of all the words that follow that key word in the original word list.
def createNextwordsPractice(wordlist):
    nextwordsPractice = {}
    for i, word in enumerate(wordlist[:-1]):
        if word not in nextwordsPractice:
            nextwordsPractice[word] = []
        nextwordsPractice[word].append(wordlist[i+1])
    return nextwordsPractice

# Function to create a dictionary of next words, similar to the previous one but with additional features.
def createNextwords(wordlist):
    nextwords = {'%': [wordlist[0]]}
    for i, word in enumerate(wordlist[:-1]):
        if word.endswith('.'):
            if '%' not in nextwords:
                nextwords['%'] = []
            nextwords['%'].append(wordlist[i+1])
        else:
            if word not in nextwords:
                nextwords[word] = []
            nextwords[word].append(wordlist[i+1])
    return nextwords

# Function to generate a play based on the given next words dictionary and the number of words specified.
def makePlay(nextwords, n):
    newplay = ''
    currentword = '%'
    for _ in range(n):
        newplay += currentword + ' '
        if currentword.endswith('.'):
            newplay += '\n'
            currentword = '%'
        else:
            currentword = choice(nextwords[currentword])
    return newplay

# Load Shakespeare's plays from URL
url = 'https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt'
text = urlopen(url).read().decode('utf-8')
shakespeare_words = text.split()[1651:]

# Generate a 1000-word Shakespearean play
nextwords = createNextwords(shakespeare_words)
newplay = makePlay(nextwords, 1000)
print(newplay)

def fibonacci(n, counter=[0]):
    counter[0] += 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1, counter) + fibonacci(n - 2, counter)

# Counting the number of times the Fibonacci function gets called for Fibonacci(17)
counter = [0]
fibonacci_17 = fibonacci(17, counter)
print("Number of times Fibonacci function gets called for Fibonacci(17): ", counter[0])
```
