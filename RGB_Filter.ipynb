```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests

def red_filter(image):
    red_only = image.copy()
    red_only[:, :, 1] = 0
    red_only[:, :, 2] = 0
    return red_only

def green_filter(image):
    green_only = image.copy()
    green_only[:, :, 0] = 0
    green_only[:, :, 2] = 0
    return green_only

def blue_filter(image):
    blue_only = image.copy()
    blue_only[:, :, 0] = 0
    blue_only[:, :, 1] = 0
    return blue_only

def red_green(image):
    red_green_only = image.copy()
    red_green_only[:, :, 2] = 0
    return red_green_only

def red_blue(image):
    red_blue_only = image.copy()
    red_blue_only[:, :, 1] = 0
    return red_blue_only

def green_blue(image):
    green_blue_only = image.copy()
    green_blue_only[:, :, 0] = 0
    return green_blue_only

def RGB_Filter():
    url = input("Enter the URL of the .jpg file: ")
    image = np.asarray(Image.open(requests.get(url, stream=True).raw))

    plt.figure(figsize=(20, 10))

    plt.subplot(2, 4, 1)
    plt.imshow(image)
    plt.title("Original Image")

    plt.subplot(2, 4, 2)
    plt.imshow(red_filter(image))
    plt.title("Red Filtered")

    plt.subplot(2, 4, 3)
    plt.imshow(green_filter(image))
    plt.title("Green Filtered")

    plt.subplot(2, 4, 4)
    plt.imshow(blue_filter(image))
    plt.title("Blue Filtered")

    plt.subplot(2, 4, 5)
    plt.imshow(red_green(image))
    plt.title("Red and Green Only")

    plt.subplot(2, 4, 6)
    plt.imshow(red_blue(image))
    plt.title("Red and Blue Only")

    plt.subplot(2, 4, 7)
    plt.imshow(green_blue(image))
    plt.title("Green and Blue Only")

    plt.show()

    # Verify if the original image is identical to the sum of the filtered images
    original_image = red_filter(image) + green_filter(image) + blue_filter(image)
    is_identical = np.array_equal(image, original_image)
    if is_identical:
        print("The original image is identical to the sum of the filtered images.")
    else:
        print("The original image is not identical to the sum of the filtered images.")

RGB_Filter()
```
